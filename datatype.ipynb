```python
d = [3,['3','4','5'],5,5, [4,5,6]]
```

```python
d
```

```python
for i in range(len(d)):
    print(f'List element {i} is {d[i]} and is of type {type(d[i])}')
```

```python
for k,v in enumerate(d):
    print(f'List element {k} is {v} and is of type {type(v)}')
```

```python
for k,v in enumerate(d):
    t = str(type(v))[8:-2]
    print(f'List element {k} is {v} and is of type {t}')
```

```python
for k,v in enumerate(d):
    t = str(type(v))[8:]
    apost = t.find("'")
    print(f'List element {k} is {v} and is of type {t[:apost]}')
```
