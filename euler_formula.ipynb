```python
def euler_from_cos_sin():
    real = eval(input('cosine part: '))
    img = eval(input('sine part: '))
    
    # print(real,img)
    m = np.sqrt(real**2 + img **2)
    k = np.arctan2(img,real)
    plt.polar([0,k],[0,m])
    plt.title('me$^{\\phi}$, m=%g, $\\phi$= %g'%(m,k))
    plt.thetagrids([0,45,130,200,222.2 ])
    plt.show()
    
    
```

```python
nvects = 200

ms = np.random.rand(nvects)
ks = np.random.rand(nvects)*2*np.pi

for i in range(0,nvects):
    r = np.random.rand()
    if r <- .4:
        clr = [1,.2, .7]
    elif r > .4 and r < .8:
        clr = [.7,.2,1]
    else:
        clr = [0,1,0]
    
    plt.polar([0,ks[i]],[0,ms[i]],color=clr)
plt.axis('off')
plt.show()
```

```python
# x[n+1] = r*x[n]*(1-x[n])
# import IPython
# %matplotlib notebook
N = 500
x = .5 + np.zeros(N)

endcap = np.arange(round(N*.9), N)

rs = np.linspace(1,4,1000)
for ri in range(len(rs)):
    for n in range(N-1):
        x[n+1] = rs[ri]*x[n]*(1-x[n]) + .2*np.sin(2*np.pi*x[n])**2
    
    u = np.unique(x[endcap])
    r = rs[ri]*np.ones(len(u))
    plt.plot(r,u,'.','.', markersize=1, color=[(np.sin(ri/len(rs)/2)+1)/2,1-ri/len(rs),.5])

plt.show()
```
