```python
def quadeq(a,b,c):
    out = np.zeros(2,dtype=complex)
    out[0] = (-b + np.lib.scimath.sqrt(b**2 - 4*a*c)) / (2*a)
    out[1] = (-b - np.lib.scimath.sqrt(b**2 - 4*a*c)) / (2*a)
    return out
```

```python
quadeq(2,7,5)
```

```python
a = 1
b = range(-5, 6)
c = range(-2, 11)
M = np.zeros((len(b),len(c)),dtype=complex)
```

```python
M
```

```python
for bi in range(0, len(b)):
    for ci in range(0, len(c)):
        M[bi,ci] = quadeq(a,b[bi],c[ci])[0]
M
```

```python
plt.subplot(1,3,1)
plt.imshow(np.real(M),extent=[c[0],c[-1], b[0],b[-1]])
# plt.xlabel('c')
# plt.ylabel('b')
# plt.axis('off')
# plt.colorbar()
# plt.legend()
plt.title('real part')

plt.subplot(1,3,2)
plt.imshow(np.imag(M),extent=[c[0],c[-1], b[0],b[-1]])
# plt.xlabel('c')
# plt.ylabel('b')
plt.axis('off')
# plt.colorbar()
# plt.legend()
plt.title('imag part')


plt.subplot(1,3,3)
plt.imshow(np.absolute(M),extent=[c[0],c[-1], b[0],b[-1]])
# plt.xlabel('c')
# plt.ylabel('b')
plt.axis('off')
# plt.colorbar()
# plt.legend()
plt.title('magnitude')

plt.show()
```
