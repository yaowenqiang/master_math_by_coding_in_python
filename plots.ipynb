```python
import sympy as sym
```

```python
x = sym.symbols('x')
y = x**2 - 3 * x

for pnti in range(-11, 12):
    plt.plot(pnti, y.subs(x,pnti), 'o')

plt.xlabel('x')
plt.ylabel('f(x) %s' %y)
plt.show()
```

y = mx + b

```python
x = [-5,5]
m = -2
b = -1
# y = m*x +b
y = [0,0]
for i in range(0,len(x)):
    y[i] = m*x[i] + b
plt.plot(x,y, label='y=%sx+%s'%(m,b))
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
x = [-5,5]
m = -2
b = -1
# y = m*x +b
# y = [0,0]
# for i in range(0,len(x)):
#     y[i] = m*x[i] + b
y = m*np.array(x) + b
plt.plot(x,y, label='y=%sx+%s'%(m,b))
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.legend()
plt.show()
```

```python
x = [-5,5]
m = [.7,-1.25]
b = [-2,.75]
for i in range(0, len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y, label='y=%sx+%s'%(m[i],b[i]))
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.legend()
plt.show()
```
