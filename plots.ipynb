y = mx + b

```python
x = [-5,5]
m = -2
b = -1
# y = m*x +b
y = [0,0]
for i in range(0,len(x)):
    y[i] = m*x[i] + b
plt.plot(x,y, label='y=%sx+%s'%(m,b))
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')
plt.legend()
plt.show()
```

```python
# soft coding
n = 10
c = np.zeros([n,n])
```

```python
for i in range(0,n):
    for j in range(0,n):
        c[i,j] = (-1)**(i+j)
```

```python
c
```

```python
plt.set_cmap('gray')
plt.tick_params(labelleft=False,labelbottom=False)
plt.imshow(c)
```

```python
from  matplotlib.patches import Polygon
```

```python
y = np.array([[1,1],[2,3],[3,1]])
```

```python
y
```

```python
p = Polygon(y,facecolor='m', alpha=.3)
y1 = np.array([[2,2],[2.5,4],[3.5,1]])
p1 = Polygon(y1,facecolor='w', alpha=.3,edgecolor='k')

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```

```python
x = np.linspace(-2,2,101)
f = -x**2
pnts = np.vstack((x,f)).T


# print(np.shape(pnts))
p = Polygon(pnts, facecolor='g',alpha=.2)
fig, ax = plt.subplots()

y = np.array([[-.5,-4],[-.5, -2.5],[.5, -2.5],[.5,-4]])
p2 = Polygon(y, facecolor='k')
ax.add_patch(p)
ax.add_patch(p2)
plt.plot(x,f,'k')
plt.plot([-2,2],[-4,-4],'k')
plt.axis('off')

plt.savefig('nicefigure.png')
plt.show()
```
