```python
def plot_complex_numbers(complex_numbers, labels=None, title="复平面上的点"):
    """
    在复平面上绘制复数
    """
    # 提取实部和虚部
    real_parts = [z.real for z in complex_numbers]
    imag_parts = [z.imag for z in complex_numbers]
    
    # 创建图形
    plt.figure(figsize=(8, 8))
    
    # 绘制复数点
    plt.scatter(real_parts, imag_parts, color='blue', s=100, zorder=5)
    
    # 添加标签
    if labels is None:
        labels = [f"{z:.2f}" for z in complex_numbers]
    
    for i, (x, y, label) in enumerate(zip(real_parts, imag_parts, labels)):
        plt.text(x, y + 0.1, label, fontsize=12, ha='center', 
                bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))
    
    # 绘制坐标轴
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    # 设置坐标轴范围和标签
    max_val = max(max(abs(np.array(real_parts))), max(abs(np.array(imag_parts)))) + 1
    plt.xlim(-max_val, max_val)
    plt.ylim(-max_val, max_val)
    
    plt.xlabel("real (Re)")
    plt.ylabel("image (Im)")
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.gca().set_aspect('equal', adjustable='box')
    
    # 添加单位圆
    theta = np.linspace(0, 2*np.pi, 100)
    plt.plot(np.cos(theta), np.sin(theta), 'r--', alpha=0.5, label='单位圆')
    plt.legend()
    
    plt.show()
```

```python
z1 = 4+5j
z2 = 3+2j
```

```python
numbers = [1+1j, 2-1j, -1+2j, -2-2j, 3j, -3j, 2, -2]
plot_complex_numbers(numbers, title="")
```
