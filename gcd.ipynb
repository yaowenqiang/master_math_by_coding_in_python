```python
import numpy as np
```

```python
m = 10
n = 15

gcd_mat = np.zeros([n,m]) + 99
```

```python
gcd_mat
```

```python
for row_i in range(0,n):
    for col_i in range(0,m):
        gcd_mat[row_i][col_i] = math.gcd(row_i+1,col_i+1)
```

```python
sym.sympify(gcd_mat)
```
