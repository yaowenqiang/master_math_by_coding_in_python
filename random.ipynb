```python
desired_mean = 15
desired_std = 4.3
nums = np.random.randn(1000) 
nums = nums - np.mean(nums)
# nums = nums + desired_mean
nums = nums /np.std(nums)
nums = nums*desired_std + desired_mean
plt.subplot(121)
plt.plot(nums,'s')
plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' %(np.mean(nums),np.std(nums)))
```

```python
def angle_convert_plot():
    angle = eval(input('what angle to convert?'))
    unit = input('which unit do you input(radian or degree)?')
    
    if unit[0].lower() == 'r':
        rad = angle
        degree = np.rad2deg(angle)
    elif unit[0].lower() == 'd':
        deg = angle
        rad = np.deg2rad(angle) 
    else:
        raise ValueError("Unknown unit!")
    
    deg = deg % 360
    rad = rad % (2 *np.pi)
    plt.plot([0,1], [0,0], 'r',linewidth=2)
    plt.plot([0,np.cos(rad)], [0,np.sin(rad)],'r', linewidth=3)
    plt.axis('square')
    plt.axis([-1,1,-1,1])
    plt.grid()
    plt.title('Angle is %s$^o$ or %s rad. '%(deg,rad))
    plt.show()
    # print(angle,rad)
```

```python
angle_convert_plot()
```
