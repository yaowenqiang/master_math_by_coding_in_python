```python
a = np.arange(1,6)
suma = np.sum(a)
sumb = np.sum(a**2)
anwser1 = suma / sumb
anwser2 = 1 / suma
```

```python
print('Option 1 = %s' % (anwser1))
print('Option 2 = %s' % (anwser2))
```

```python
print('Option 1 = %s\nOption 2 = %s' % (anwser1,anwser2))
```

```python
product_a = np.prod(a)
product_b = np.prod(a**2)
anwser1 = product_a / product_b
anwser2 = 1 / product_a
print('Option 1 = %s\nOption 2 = %s' % (anwser1,anwser2))
```
