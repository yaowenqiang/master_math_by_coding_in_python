## Law of exponents

```python
x,y,z = sym.symbols('x,y,z')
```

```python
Math('x^yx^z = x^{y+z}')
```

```python
Math('\\frac{x^y}{x^z} = x^yx^{-z} = x^{y-z}')
```

```python
Math('x^yy^z = x^yy^z')
```

```python
ex = x**y * x**z
sym.simplify(ex)
```

```python
exp1 = x**y * x ** z
exp2 = x**y / x ** z
exp3 = x **y * y ** z
display(Math('%s = %s' %(sym.latex(exp1), sym.latex(sym.simplify(ex1)))))
```

```python
display(Math('%s = %s' %(sym.latex(exp2), sym.latex(sym.simplify(ex2)))))
```

```python
display(Math('%s = %s' %(sym.latex(exp3), sym.latex(sym.simplify(ex3)))))
```
