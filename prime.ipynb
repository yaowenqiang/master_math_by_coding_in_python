```python
for i in range(2, 51):
    factor_dict = sym.factorint(i)
    keys= list(factor_dict.keys())
    if len(keys) == 1 and factor_dict[keys[0]] == 1:
        print('%g is a prime number' % i)
    else:
        print('%g is a composite number with prime factors: %s' %(i, keys))
```
