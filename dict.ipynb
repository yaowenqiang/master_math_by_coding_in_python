```python
import sympy as sym
from IPython.display import display,Math
```

```python
x,y = sym.symbols('x,y')
```

```python
D = dict(equ_x = [4*x -6, x**2 -9], equ_y = [sym.sin(y)])
```

```python
D
```

```python
for key_i in D:
    print('solutions for equations involing ' + key_i[-1] + ':')
    for eqi in D[key_i]:
        left_part = sym.latex(eqi) + ' = 0'
        middle_part = ' \\quad \\rightarrow \\quad ' + key_i[-1] + '='
        right_part = sym.latex(sym.solve(eqi))
        display(Math(' \\quad\\quad ' + left_part + middle_part + right_part))
```
