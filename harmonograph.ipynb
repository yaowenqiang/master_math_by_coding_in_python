```python
def powers(x,y):
    display(Math("%g^%g = %g"%(x,y, x ** y)))
    # return x ** y

def division(x,y):
    display(Math("\\frac{%g}{%g} = %g"%(x,y,x / y)))
    # return x / y
```

```python
def main():
    x = int(input("Input x: "))
    y = int(input("Input y: "))
    display(Math('\\text{Press "1" to compute }%g^%g \\text{ or press "2" to compute }\\frac{%g}{%g}'%(x,y,x,y)))
    switch = input(" ")
    result = None
    if switch == '1':
        result = powers(x,y)
    elif switch == '2':
        result = division(x,y)
    else:
        print("Invalid selection!") 
    if result is not None:
        print(result)

   
```

```python
main()
```

```python
def guess_the_number():
    num_2_guess = random.randint(1,101)
    user_guess = int(input("Guess a number between 1 and 100:    "))
    while user_guess != num_2_guess:
        if user_guess < num_2_guess:
            print("Guess higher!")
        elif user_guess > num_2_guess:
            print("Guess lower!")
        else:
            break
        user_guess = int(input("Guess again:     "))
    
    print("YES!!! you guessed correctly!")
            
```
