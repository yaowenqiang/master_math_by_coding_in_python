```python
import sympy as sym
from IPython.display import display,Math
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
#%matplotlib %notebook
```

```python
x = sym.symbols('x')
fx = x**3
limit_point = 1.5
limit = sym.limit(fx, x, limit_point)
display(Math('\\lim_{x\\to %g} %s = %g'%(limit_point, sym.latex(fx), limit)))
```

```python
fxx = sym.lambdify(x,fx)
fxx
```

```python
fxx(0)
```

```python
fxx(1.5)
```

```python
xx = np.linspace(-5,5,200)
plt.plot(xx, fxx(xx))
plt.plot(limit_point, limit,'ro')
```

```python
fx = (x**2)/(x-2)
fxx = sym.lambdify(x,fx)
xx = np.linspace(1,3,102)
lim_pnt = 2
lim = sym.limit(fx, x, lim_pnt,dir='-')
display(Math('\\lim_{x\\to %g^-} %s = %g'%(lim_pnt, sym.latex(fx), lim)))
```

```python
plt.plot(xx,fxx(xx))
```

```python
x = sym.symbols('x')
fx = sym.sqrt(x+1)* sym.exp(-x)
gx = sym.cos(x + sym.sin(x))
fxx = sym.lambdify(x,fx)
gxx = sym.lambdify(x, gx)
```

```python
xx = np.linspace(0,10, 100)
plt.plot(xx, fxx(xx), label='f(x)')
plt.plot(xx, gxx(xx), label='g(x)')
plt.legend()
plt.show()
```

```python
limit_pnt = 5
lim_fx = sym.limit(fx, x, limit_pnt)
lim_gx = sym.limit(gx, x, limit_pnt)
```

```python
hx = fx / gx
lim_fgx = sym.limit(hx, x, limit_pnt)

display(Math('\\frac{\\lim_{x \\to %g} f(x)}{\\lim_{x \\to %g} g(x)} = \\frac{%g}{%g} = %g' \
             %(limit_pnt, limit_pnt, lim_fx, lim_gx, lim_fx/lim_gx)))
display(Math('\\lim_{x\\to %g} \\frac{f(x)}{g(x)} = %g'%(limit_pnt, lim_fgx)))
```

```python
from sympy.abc import x
piece1 = 0
piece2 = -2*x
piece3 = x**3/10

fx = sym.Piecewise((piece1, x<0),(piece2,(x>=0) & (x < 3)),(piece3, x>=3))
```
