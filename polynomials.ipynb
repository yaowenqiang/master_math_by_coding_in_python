```python
polys = [sym.poly(2*x + x **2), sym.Poly(-x**3 + 4*x), sym.Poly(x**5 - x**4 + x/4 + 4)]
```

```python
polys
```

```python
for p in polys:
    # display(Math(sym.latex(p)))
    # print(p.degree())
    if p.degree() %2 == 0:
        print('The degree of %s is even, and the coefficients sum to %g.'%(p.as_expr(),sum(p.coeffs())))
    else:
        print('The degree of %s is odd, and there are %g coeffeients.'%(p.as_expr(), len(p.coeffs())))
```

```python
exprs = [x**2+4*x+3, 2*y**2-1,3*y**2+12*y]
for e in exprs:
    fac = sym.factor(e)
    str_fac = str(fac)
    if str_fac.find('(') != -1:
        display(Math('%s \\quad \\Rightarrow \\quad %s '%(sym.latex(e), sym.latex(fac))))
    else:
        display(Math('%s \\quad \\Rightarrow \\quad \\text{ not factorable} '%(sym.latex(e))))
```
