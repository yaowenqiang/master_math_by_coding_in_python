```python
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
```

```python
expr = 2*x + 4 - 9
sym.solve(expr)
```

```python
display(Math('\\text{The solution is } %s \\text{ is x=} %g' %(sym.latex(expr), sym.solve(expr)[0])))
```

```python
eq
```
