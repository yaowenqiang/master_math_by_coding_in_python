```python
x,y = sym.symbols('x,y')
```

```python
e1 = 2*x + x*(4-6*x) + x
```

```python
e1
```

```python
e1.simplify()
```

```python
e1.simplify()
```

```python
e2 = -x *(2/x + 4 / (x**2)) + (4 + x) / (4 *x)
```

```python
e2
```

```python
e3 = (x + 3) * (x -3) * x * (1 /( 9 *x))
```

```python
e3
```

```python
exports = [e1,e2,e3]
```

```python
exports
```

```python
display(Math("%s \\quad \\longleftrightarrow \\quad %s"%(sym.latex(e1), sym.latex(sym.expand(e1)))))
```

```python
for expri in exports:
    display(Math("%s \\quad \\longleftrightarrow \\quad %s"%(sym.latex(expri), sym.latex(sym.expand(expri)))))
```
